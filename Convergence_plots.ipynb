```python
import matplotlib.pyplot as plt
```

```python
f = open("convergence-clean.txt","r")
contents = f.readlines()
```

```python
convergence = {}
for line in contents:
    line_arr = line.split()
    # New model
    if '-' in line_arr[0]:
        continue
    elif '|' in line_arr:
        convergence[name]['iteration'].append(int(line_arr[0]))
        convergence[name]['train'].append(float(line_arr[2]))
        convergence[name]['validation'].append(float(line_arr[4]))
    else:
        name = line_arr[0]
        convergence[name] = {'iteration': [], 'train': [], 'validation': []}
models = convergence.keys()
```

```python
models = convergence.keys()
titles = ['L=10', 'L=100', 'L=500']
fig = plt.figure(figsize=(10,4))
for i, model in enumerate(models):
    if i == 2:
        plt.title('Convergence rates for ranking quality, Ndcg@5\n')
    plt.subplot2grid((1,3), (0,i))
    plt.plot(convergence[model]['iteration'], convergence[model]['train'], label=titles[i] + ', training')
    plt.plot(convergence[model]['iteration'], convergence[model]['validation'], label=titles[i] + ', validation')
    plt.legend(loc='lower right')
plt.show()
```
